```python
%sql
    
CREATE or REPLACE TABLE sales_project_db.Sales_category
AS 
SELECT Category,`Sub_Category`,round(sum(Total_sales),4) as Total_Sales, count(Quantity) as Number_of_Products_sold 
from sales_project_db.silver_sales
group by Category,`Sub_Category`
sort by Category;
```

```python
%sql
CREATE OR REPLACE TABLE sales_project_db.gold_delivery_performance AS
SELECT 
    delivery_type,
    COUNT(*) as total_orders,
    round(AVG(delivery_days),2) as avg_delivery_days
FROM sales_project_db.silver_sales
GROUP BY delivery_type;
```

```python
%sql
CREATE OR REPLACE TABLE sales_project_db.gold_date_wise_sales AS
SELECT 
    
    year(Order_Date) as order_year,
    month(Order_Date) as order_month,
    date(Order_Date) as order_date,
    weekofyear(Order_Date) as order_week,
    SUM(Sales) as total_sales,
    SUM(Total_sales) as total_revenue
FROM sales_project_db.silver_sales
GROUP BY 
    year(Order_Date),
    month(Order_Date),
    weekofyear(Order_Date),
    date(Order_Date)
SORT BY order_year,order_month,order_week;
```

```python
%sql
CREATE OR REPLACE TABLE sales_project_db.gold_customer_analysis AS
SELECT 
    Customer_ID,
    Customer_Name,
    COUNT(DISTINCT Order_ID) as total_orders,
    ROUND(SUM(Sales),4) as total_spent,
    COUNT(DISTINCT year(Order_Date)) as active_years,
    CASE 
    WHEN total_orders >= 10 THEN 'Regular'
    ELSE 'Occasional'
END as customer_type
FROM sales_project_db.silver_sales
GROUP BY Customer_ID, Customer_Name
SORT BY total_orders DESC;
```

```python
%sql
CREATE OR REPLACE TABLE sales_project_db.gold_product_performance AS
SELECT 
    Product_ID,
    Category,
    Sub_Category,
    COUNT(*) as times_ordered,
    SUM(Quantity) as total_quantity,
    SUM(Sales) as total_sales
FROM sales_project_db.silver_sales
GROUP BY Product_ID, Category, Sub_Category;
```

```python
%sql
CREATE OR REPLACE TABLE sales_project_db.gold_customer_country AS
SELECT 
    Country,
    City,
    COUNT(DISTINCT Customer_ID) as total_customers,
    SUM(Sales) as total_sales
FROM sales_project_db.silver_sales
GROUP BY Country, City;
```

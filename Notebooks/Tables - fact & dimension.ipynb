Creating fact and dimension table - 

Orders - fact table :  columns for orders, dates, sales data  
Products - dimension table :  columns for product id, categories  
Customers - dimension table : customer name, id, country, city, state, region 

```python
%sql
CREATE OR REPLACE TABLE sales_project_db.orders AS
SELECT 
    Order_ID,
    Order_Date,
    Ship_Date,
    Customer_ID,
    Product_ID,
    Sales,
    Quantity,
    Discount,
    Residue,
    Total_sales,
    delivery_days,
    delivery_type
FROM sales_project_db.silver_sales;
```

```python
%sql
CREATE OR REPLACE TABLE sales_project_db.customers AS
SELECT DISTINCT
    Customer_ID,
    Customer_Name,
    Segment,
    Country,
    City,
    State,
    Region
FROM sales_project_db.silver_sales;
```

```python
%sql
CREATE OR REPLACE TABLE sales_project_db.products AS
SELECT DISTINCT
    Product_ID,
    Category,
    Sub_Category
FROM sales_project_db.silver_sales;
```

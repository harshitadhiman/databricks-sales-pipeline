```python
df = spark.table("workspace.default.samplesuperstore_1")
```

```python
df_Clean = df.toDF(*[col.replace(" ", "_") for col in df.columns]).withColumnsRenamed({"Country/Region":"Country","State/Province":"State"})

# delta tables not support blank spaces in headers, replaced by '_' 
df_Clean.display()
```

```python
df_Clean.write.format("delta").mode("overwrite").option("overwriteSchema", "true").saveAsTable("workspace.default.bronze_sales")
#redirect to either a Unity Catalog table using saveAsTable() or a Volume path
```
